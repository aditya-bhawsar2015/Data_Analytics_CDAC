```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('heart.csv')
ages = df.groupby("age").median().reset_index()

ages.plot.line("age","chol", marker ='o')
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(ages["age"], ages['chol'], ls =":", lw = 1.7)
ax.set_xlabel("Age")
ax.set_ylabel("Cholestrol")
plt.show()
```

```python
sns.lineplot(x='age', y='chol', data = ages, ls = ':', lw = 1.7)
plt.xlabel('Age')
plt.show()
```
